```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cf
import pandas as pd
```

read in the data from the dangerzone spreadsheet, and get the columns for Average Latitude and Average Longitude and convert them to lists using the tolist function 

```python
irmadata=pd.read_excel('DangerZone.xlsx')
irmalats=irmadata['Average Latitude'].tolist()
irmalong=irmadata['Average Longitude'].tolist()
```

create a list to set the extent of the projection, use the values -50, -90, 15, 50
then create the axes object using the axes method of the pyplot module from matplotlib as before
set the stock image, and the extent, then use the coastlines method to set the resolution='50m'


```python
extent=[-50,-90,15,50]         #setting extent of the longitudes and latitudes to be covered in the map
ax=plt.axes(projection=ccrs.PlateCarree())   #projection type of the map
ax.stock_img()
ax.add_feature(cf.COASTLINE)
ax.set_title('Position of Ships')  
ax.set_extent(extent)              #plotting the extent of your map
ax.coastlines(resolution='50m')    #Setting resolution to 50m
ax.gridlines()                     #add gridlines
ax.legend()
#Plotting the ship positions (longitudes, latitudes, color/circle and label for the legend)
ax.plot(-85,17,'ro',label='ship 1')
ax.plot(-78,23,'yo', label='ship 2')
ax.plot(-68,18.5,'ko',label='ship 3')
ax.plot(-72,20,'go',label='ship 4')
ax.plot(-73,17.7,'bo',label='ship 5')
ax.plot(-65,23,'mo',label='ship 6')
ax.plot(irmalong,irmalats)    #plotting the required longitudes and latitudes

plt.show()
```
